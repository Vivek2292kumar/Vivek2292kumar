By examining how different minimum night requirements and the frequency of availability impact booking rates, the analysis aims to provide recommendations for optimizing these parameters. ### Variables Description

*Answer*

**id** : Unique identifier for each listing and property

**name**:  Name or title of the listing, typically provided by the host.

**host_id**: Unique identifier for the host who manages the listing.

**host_name**: Name of the host managing the listing.

**neighbourhood_group**: Broad area or borough where the listing is located (e.g., Manhattan, Brooklyn).

**neighbourhood**: Specific neighborhood within the neighbourhood_group where the listing is located.

**latitude**: Latitude coordinate of the listings location, used for mapping.

**longitude**: Longitude coordinate of the listings location, used for mapping.

**room_type**: Type of room being offered (e.g., Entire home/apt, Private room, Shared room).

**price**: Price per night for the listing, in the local currency (e.g., USD).

**minimum_nights**: Minimum number of nights required for a booking.

**number_of_reviews**: Total number of reviews received by the listing.

**last_review**: Date of the most recent review for the listing.

**reviews_per_month**: Average number of reviews the listing receives per month.

**calculated_host_listings_count**: Total number of listings the host manages, reflecting their hosting experience.

**availability_365**: Number of days the listing is available for booking in a year (365 days).

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# chnge 'last_review' column to datetime from object
df['last_review'] = pd.to_datetime(df['last_review'])
# impute missing value
# impute missing values with 0 in columns like
df['last_review'] = df['last_review'].fillna(df['last_review'].max())
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
# remove unnecesarry columns
df.drop(['name','host_name'] , axis = 1 , inplace = True)
````
